```python
import pandas as pd
import numpy as np
```

```python
# 1. Crear arreglo con los datos de las mascotas
mascotas = np.array([['Bella', 'Labrador', 'Cafe', 56, 24, '2013-07-01'],
                     ['Charlie', 'Poodle', 'Negro', 43, 24, '2016-09-16'], 
                     ['Lucy', 'Chow chow', 'Cafe', 46, 24, '2014-08-25'],
                     ['Cooper', 'Schnauzer', 'Gris', 49, 17, '2011-12-11'],
                     ['Bernie', 'San Bernardo', 'Blanco', 77, 74, '2018-02-27']])

mascotas
```

```python
# 2. Crear dataframe <df>
df = pd.DataFrame(data=mascotas, columns=['Nombre', 'Raza', 'Color', 'Altura_cm', 'Peso_kg', 'Fecha_Nac'])
df
```

```python
# 3. Hacer otro df con datos de perros cuya fecha de nacimiento sea previa a 01 ene 2025
df2 = df[df['Fecha_Nac'] < '2015-01-01']
df2
```

```python
# 4. Cambiar el formato de caracter a numero
df['Altura_cm'] = df['Altura_cm'].astype(int)
df['Peso_kg'] = df['Peso_kg'].astype(int)
```

```python
# 5. Filtrar los perros cuya altura sea mayor a 50
df[df['Altura_cm'] > 50]
```

```python
# 6. Agregar columna de IMM (peso/altura)*2
df['IMM'] = df['Peso_kg'] / (df['Altura_cm'] / 100) ** 2
df
```

```python
# 7. Hacer otro df con datos de IMM ordenados
df3 = df[['Nombre', 'Raza', 'IMM']]
df3 = df3.sort_values('IMM', ascending=True)
df3
```

```python
# 8. Crear arreglo con los datos adicionales de las mascotas
mascotas = np.array([['Bella', 'Labrador', 'Cafe', 56, 24, '2013-07-01'],
                     ['Charlie', 'Poodle', 'Negro', 43, 24, '2016-09-16'], 
                     ['Lucy', 'Chow chow', 'Cafe', 46, 24, '2014-08-25'],
                     ['Cooper', 'Schnauzer', 'Gris', 49, 17, '2011-12-11'],
                     ['Bernie', 'San Bernardo', 'Blanco', 77, 74, '2018-02-27'],
                     ['Max', 'Chow chow', 'Blanco', 45, 28, '2019-03-27'],
                     ['Red', 'Schnauzer', 'Negro', 45, 15, '2017-09-18']])
mascotas
```

```python
# 9. Crear dataframe y modificar tipo de datos de las columnas Peso y Altura
df = pd.DataFrame(data=mascotas, columns=['Nombre', 'Raza', 'Color', 'Altura_cm', 'Peso_kg', 'Fecha_Nac'])
df['Peso_kg'] = df['Peso_kg'].astype(int)
df['Altura_cm'] = df['Altura_cm'].astype(int)
df
```

```python
# 10. Calcular media / promedio 

df['Peso_kg'].mean()
```

```python
# 11. Obtener la fecha de nacimiento más antigua
df['Fecha_Nac'].min()
```

```python
df['Altura_cm'].cumsum()
```

```python
# 12. Realizar conteo por raza
df['Raza'].value_counts()
```

```python
# 13. FUNCIONES DE AGRUPAMIENTO
# Calcular el valor promedio del peso por cada color

df.groupby('Color')['Peso_kg'].mean()
```

```python
df.groupby('Color')['Peso_kg'].agg(["min", "max", "sum"])
```

```python
# Agrupar por color y raza en base a la media del peso

df.groupby(['Color', 'Raza'])['Peso_kg'].mean()
```

```python
# Uso de tabla pivote

df.pivot_table(values = 'Peso_kg', index = 'Color', aggfunc = ["mean", "median"])
```

```python
# Obtener promedio para 2 variables
# Calcula la media por omisión

df.pivot_table(values = 'Peso_kg', index = 'Color', columns = 'Raza', fill_value = 0)
```

```python
# Calcular promedio por columna o fila

df.pivot_table(values = 'Peso_kg', index = 'Color', columns = 'Raza', fill_value = 0, margins = True, aggfunc = "mean")
```
