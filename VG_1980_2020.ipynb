```python
# Al observar que no coincide el número de registros en la columna 'Year' con las demás, se realizó una búsqueda de valores nulos.
print("df_vg:\n", df_vg.isna().sum())
```

```python
# Se modifico el tipo de dato de la columna 'Year' para mayor claridad.

df_vg['Year'] = df_vg['Year'].astype('Int64')
```

```python
# 2. Totales:
# 2.1. Total de videojuegos analizados -> Se analizaron 11,493 vg

total_vg = df_vg['Name'].nunique()
total_vg
```

```python
# 2.2. Total de géneros analizados -> Se analizaron 12 géneros de vg

df_vg.groupby('Genre').size().reset_index(name='Total').sort_values(by='Total', ascending=False)
```

```python
# 2.3. Total de plataformas analizadas -> Se analizaron 31 plataformas de vg
df_vg.groupby('Platform').size().reset_index(name='Total').sort_values(by='Total', ascending=False)
```

```python
# 3. Ventas: 
# 3.1. Total de ventas por año
df_sales = df_vg.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].sum().reset_index().sort_values(by='Global_Sales', ascending=False)
df_sales
```

```python
fig = px.bar(df_sales, x='Year', y='Global_Sales', title='Ventas por año', text_auto=True)
fig.update_xaxes(title='Año', dtick=1)
fig.update_yaxes(title='Ventas globales')
fig.show()
```

```python
fig = px.bar(df_sales, x='Year', y=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], title='Ventas por región a lo largo del tiempo', labels={'value': 'Ventas globales', 'Year': 'Año'})
fig.show()
```

```python
# 3.2. Juegos más vendidos por año
df_sales_vg = df_vg.groupby(['Year', 'Name'])[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].sum().reset_index()
df_sales_vg
```

```python
df_top_games = (df_sales_vg.loc[df_sales_vg.groupby('Year')['Global_Sales'].idxmax()].sort_values(by='Year').reset_index(drop=True))
df_top_games
```

```python
fig = px.bar(df_top_games, x='Year', y='Global_Sales', text='Name', title='Videojuegos más vendidos a tráves del tiempo', labels={'Global_Sales': 'Ventas globales', 'Year': 'Año'})
fig.update_traces(textposition='inside', textangle=-90)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# 3.3 Plataformas con más ingresos por año
# Todos estos números desglosados por ingreso: NA, EU, JP, Otros y el Global

df_sales_platform = df_vg.groupby(['Year', 'Platform'])[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].sum().reset_index().sort_values(by='Global_Sales', ascending=False)
df_sales_platform
```

```python
fig = px.bar(df_sales_platform, x='Platform', y='Global_Sales', color='Year', title='Ventas entre plataformas por año')
fig.update_xaxes(title='Plataforma', dtick=1)
fig.update_yaxes(title='Ventas globales')
fig.show()
```

```python
# 4. Gráfico de comparación de ventas entre géneros por año

df_sales_genre = df_vg.groupby(['Year', 'Genre'])[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].sum().reset_index().sort_values(by='Global_Sales', ascending=False)
df_sales_genre
```

```python
fig = px.bar(df_sales_genre, x='Year', y='Global_Sales', facet_col='Genre')
fig.update_yaxes(showticklabels=False)
fig.update_layout(yaxis_title='Ventas globales')
fig.update_xaxes(matches='x')

for axis in fig.layout:
    if axis.startswith("xaxis"):
        fig.layout[axis].title.text = 'Año'

fig.show()
```

```python
# 5. En los últimos 4 años: Se seleccionó el periodo de 2013 a 2016 para tener mayor número de datos para el análisis. 
# 5.1. ¿qué género ha tenido el mayor y menor crecimiento?
fig = px.strip(df_sales_genre[(df_sales_genre['Year'] >= 2013) & (df_sales_genre['Year'] <= 2016)], x ='Genre', y='Global_Sales', color='Year', title='Ventas globales por género (2013-2016)')
fig.update_layout(yaxis_title='Ventas globales')
fig.update_layout(xaxis_title='Género')
fig.show()
```

```python
# 5.2. ¿qué plataforma ha tenido el mayor y menor crecimiento? 
fig = px.strip(df_sales_platform[(df_sales_platform['Year'] >= 2013) & (df_sales_platform['Year'] <= 2016)], x ='Platform', y='Global_Sales', color='Year', title='Ventas globales por plataforma (2013-2016)')
fig.update_layout(yaxis_title='Ventas globales')
fig.update_layout(xaxis_title='Plataforma')
fig.show()
```

```python
# 5.3 ¿qué videojuego ha tenido el mayor y menor crecimiento?

vg_years = (df_sales_vg.groupby('Name')['Year'].nunique().reset_index().query('Year > 1'))
df_sales_vg_filtered = df_sales_vg[df_sales_vg['Name'].isin(vg_years['Name'])]
```

```python
df_sorted = df_sales_vg_filtered.sort_values(['Name', 'Year'])

growth = (df_sorted.groupby('Name').agg(first_year=('Year', 'first'), last_year=('Year', 'last'), first_sales=('Global_Sales', 'first'), last_sales=('Global_Sales', 'last')).assign(growth=lambda x: x['last_sales'] - x['first_sales']).reset_index())
```

```python
growth.sort_values(by='growth', ascending=False)
```

```python
# Top 5 con mayor crecimiento
top_growth = growth.sort_values('growth', ascending=False).head(5)['Name']

# Top 5 con mayor decrecimiento
low_growth = growth.sort_values('growth', ascending=True).head(5)['Name']


df_top = df_sorted[df_sorted['Name'].isin(top_growth)].copy()
df_low = df_sorted[df_sorted['Name'].isin(low_growth)].copy()

df_top['Group'] = 'Mayor crecimiento'
df_low['Group'] = 'Mayor decrecimiento'

df_compare = pd.concat([df_top, df_low])
df_compare
```

```python
fig = px.line(df_compare, x='Year', y='Global_Sales', color='Name', facet_col='Group', facet_col_wrap=2, markers=True, title='Comparación entre videojuegos con mayor crecimiento y decrecimiento')
fig.update_layout(xaxis_title='Año', yaxis_title='Ventas globales', legend_title='Videojuego')
for axis in fig.layout:
    if axis.startswith("xaxis"):
        fig.layout[axis].title.text = 'Año'
fig.show()
```
