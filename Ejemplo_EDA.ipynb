```python
# Identificar variación en los datos

ax = sns.boxplot(x=df['sqft_living'])
```

```python
# Crear un mapa de correlación

import plotly.express as px

fig = px.imshow(df.corr(numeric_only=True), text_auto=True, title='Mapa de calor de correlación', color_continuous_scale = 'viridis')
fig.show()
```

```python
# Generar gráficos exploratorios de todas las columnas

cols = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above',
'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']      
```

```python
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(18,16))
for i, column in enumerate(cols):
    sns.distplot(df[column], ax=axes[i//4,i%4], kde=False)
```

```python
# Analizar 2 variables
pd.crosstab(df['waterfront'], df['floors'])
```

```python
pd.crosstab(df['grade'], df['floors'], margins=True, margins_name='Total')
```

```python
df1 = pd.crosstab(df['grade'], df['floors'], values=df.price, aggfunc='mean', margins=True, margins_name='Total')
df1 = df1.replace(np.nan, 0)
df1
```

```python
df1.style.background_gradient(cmap = 'Blues')
```
